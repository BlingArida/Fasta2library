```python
from anarci import number
import pandas as pd
import numpy as np
import os
import json

import re
#import pysnooper

__author__ = 'yincan'
__contact__ = 'yin_can@wuxibiologics.com'

class divideScheme:
    '''
    Almost base on IMGT numbering scheme.
    The boundary of CDR2 broadens by 1 to the left and right. 
    '''
    def __init__(self, chain_type):
        self.chain_type = chain_type
        self.__getCDR()
    @staticmethod
    def convertRange(start, end):
        extent = [i for i in range(start, end + 1)]
        str_ext = [st for st in map(str, extent)]
        return str_ext
    def __getCDR(self):
        if self.chain_type.upper() == "L":
            self.cdr1 = self.convertRange(27, 38)
            self.cdr3 = self.convertRange(105, 117)
            self.fw1 = self.convertRange(1, 26)
            self.fw4 = self.convertRange(118, 127)
            #The boundary of CDR2 broadens by 1 to the left and right.
            self.cdr2 = self.convertRange(55, 66)
            self.fw2 = self.convertRange(39, 54)
            self.fw3 = self.convertRange(67, 104)
            self.span: list = [1, 127]
        elif self.chain_type.upper() == "H":
            self.cdr1 = self.convertRange(27, 38)
            self.cdr3 = self.convertRange(105, 117)
            self.fw4 = self.convertRange(118, 128)
            self.fw1 = self.convertRange(1, 26)
            #The boundary of CDR2 broadens by 1 to the left and right.
            self.cdr2 = self.convertRange(55, 66)
            self.fw2 = self.convertRange(39, 54)
            self.fw3 = self.convertRange(67, 104)
            self.span: list = [1, 128]

class Fasta2Library:
    def __init__(self, fasta, scheme='imgt'):
        self.fasta = fasta
        self.scheme = scheme        
        '''
        Numbering sequence by ANARCI's function "number"
        '''
        seqlist = self.seqS()
        data = []
        for self.seq in seqlist:
            self.numbering, self.chain_type = number(self.seq, self.scheme)
            pos_aaDic = self.outNumbering()
            seqdict = {'chain':self.chain_type,'seq':pos_aaDic}
            data.append(seqdict)
    def seqS(self):
        '''
        Merge multi-line fasta file to one-line fasta sequence.
        All the fastas consist of seqlist.
        '''
        seqlist = []
        with open(self.fasta, 'r') as f:
            seq = ''
            for line in f:
                if not line.startswith('>'):
                    seq += line.replace('\n', '').strip()
                else:
                    if seq != '':
                        seqlist.append(seq)
                    seq = ''
        return seqlist
    def outNumbering(self):
        '''
        Process on number output of numbering
        ''' 
        dict_numbering = {}
        for tmp in self.numbering:
            pos = str(tmp[0][0])
            AA = tmp[1]
            dict_numbering[pos] = AA
        return dict_numbering
```
